```python
import pandas as pd
stud=pd.read_excel(r"C:\Users\PERSONAL\Downloads\studid.xlsx")
stud
```

```python
det=pd.read_excel(r"C:\Users\PERSONAL\Downloads\STud2.xlsx")
det
```

```python
innerjoin=pd.merge(stud,det,how='inner',on='Student_Id')
innerjoin
```

```python
leftjoin=pd.merge(stud,det,how='left',on='Student_Id')
leftjoin
```

```python
rightjoin=pd.merge(stud,det,how='right',on='Student_Id')
rightjoin
```

```python
outerjoin=pd.merge(stud,det,how='outer',on='Student_Id')
outerjoin
```

```python
outerjoin.set_index('Student_Id', inplace=True)
outerjoin
```

```python
outerjoin.reset_index(inplace=True)
outerjoin
```

```python
outerjoin.to_csv('outer.csv',index=False)
outerjoin
```
